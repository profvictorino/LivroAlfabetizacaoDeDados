## Programação com Decisão

### __Exemplo 01:__ Implemente um programa com um desvio condicional simples que receba dois números e teste se a soma desses dois números é maior que 10.
### Caso o resultado desse teste seja verdadeiro, o programa deve exibir a soma, caso contrário, nenhum número será exibido.
### Quando finalizar o desvio condicional, o programa deverá imprimir a frase “Estou fora da estrutura de decisão.” antes de finalizar sua execução.

fig-94.png

```python
# Programa com desvio condicional simples
# Figura 9.4

a = int(input('Informe o valor de a:  '))
b = int(input('Informe o valor de b:  '))

x = a + b

if(x > 10):
  print ('O valor de x é: ', x)

print ('Estou fora da estrutura de decisão.')
```

### __Exemplo 02:__ Implemente um programa com um desvio condicional composto que receba dois números e teste se a soma desses dois números é maior que 10.
### Caso o resultado desse teste seja verdadeiro, o programa deve exibir o valor do primeiro número digitado, caso contrário, o programa deve exibir o valor do segundo número digitado.
### Quando finalizar o desvio condicional, o programa deverá imprimir a frase “Estou fora da estrutura de decisão.” antes de finalizar sua execução.

fig-96.png

```python
# Programa com desvio condicional composto
# Figura 9.6

a = int(input('Informe o valor de a:  '))
b = int(input('Informe o valor de b:  '))

x = a + b

if(x > 10):
    print ('O valor de a é: ', a)
else:
     print ('O valor de b é: ', b)

print ('Estou fora da estrutura de decisão.')
```

### __Exemplo 02:__ Implemente um programa utilizando laço condicional com teste no início que leia a idade de várias pessoas e no final apresente a média das idades.
### O critério de parada será a digitação de uma idade igual a zero ou negativa, ou seja, uma idade menor ou igual a zero.

fig-910.png

```python
# Programa com laço condicional com teste no início
# Figura 9.10

cont = 0
total = 0

idade = int(input('Informe uma idade: '))

while (idade > 0):
    total = total + idade
    cont = cont + 1
    idade = int(input('Informe uma idade: '))

media = total/cont

print('A média de idade é: ', media)

```
